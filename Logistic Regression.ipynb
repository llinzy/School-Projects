```python
#create a list of values for column with string structures
colsnc=[]
for k in titan2.loc[:,"Country"]:
    colsnc.append(k)
```

```python
#create a dictionary to assign values to columns with string structures
n=0
colnum={}
for l in colsnc:
    if not colnum.get(l):
        n+=1
        colnum[l]=n
```

```python
titan2.head()
```

```python
df5= pd.DataFrame({'COUNTRY':titan2.loc[:,"Country"]})
df5.replace(colnum, inplace=True)
df5.head()
```

```python
titan3=pd.concat([titan2,df5], axis=1)
titan3.head()
```
